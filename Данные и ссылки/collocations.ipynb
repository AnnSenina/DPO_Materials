```python
import nltk, codecs
from nltk.collocations import *
```

```python
#считываем файл, строки кладем в массив
corpus = []
f = codecs.open('E:\\GoogleDrive\\0_2019_Edu_CL1_DO\\Collocations\\Data\\court-V-N.csv', mode = "r", encoding = "utf-8")
for line in f:
    line = line.rstrip()
    text = [x.strip() for x in line.split(",")]
    corpus.append(text)
f.close()
print(corpus[:10])
```

```python
finder = BigramCollocationFinder.from_documents(corpus)
finder.apply_freq_filter(3)
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder.nbest(bigram_measures.pmi, 10)
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
bigrams = []
for tokens in corpus:
    finder = BigramCollocationFinder.from_words(tokens, window_size=3)
    scored = finder.score_ngrams(bigram_measures.raw_freq)
    for score in scored:
        bigrams.append(score[0])
print(bigrams[:10])
```

```python
finder2 = BigramCollocationFinder.from_words(bigrams)
finder.apply_freq_filter(3)
metrics1 = finder2.score_ngrams(bigram_measures.pmi)
res1 = sorted(metrics1, key=lambda x: x[1], reverse=True)
print ('PMI:')
print (res1[:10])
top_pmi = finder2.nbest(bigram_measures.pmi, 10)
print(metrics1)
print(top_pmi)
```

```python
metrics2 = finder2.score_ngrams(bigram_measures.chi_sq)
res2 = sorted(metrics2, key=lambda x: x[1], reverse=True)

print ('Chi-square:')
print (metrics2[:10])
```
